```python
import speech_recognition as sr
from googletrans import Translator

# Initialize recognizer and translator
recognizer = sr.Recognizer()
translator = Translator()

# Choose the target language for translation (e.g., 'fr' for French, 'es' for Spanish)
target_language = 'es'

# Function to recognize speech
def recognize_speech():
    with sr.Microphone() as source:
        print("Say something...")
        recognizer.adjust_for_ambient_noise(source)
        audio = recognizer.listen(source)
        
        try:
            # Recognize speech using Google Web Speech API
            text = recognizer.recognize_google(audio)
            print(f"Original Speech: {text}")
            return text
        except sr.UnknownValueError:
            print("Could not understand audio")
        except sr.RequestError:
            print("Error with the Google Speech Recognition service")
        return None

# Function to translate recognized speech
def translate_speech(text, target_language):
    try:
        translation = translator.translate(text, dest=target_language)
        print(f"Translated Speech: {translation.text}")
    except Exception as e:
        print(f"Error in translation: {e}")

# Main workflow
speech_text = recognize_speech()
if speech_text:
    translate_speech(speech_text, target_language)
```

```python
import speech_recognition as sr
from googletrans import Translator
from gtts import gTTS
from langdetect import detect
from pydub import AudioSegment
from pydub.playback import play
import io

recognizer = sr.Recognizer()
translator = Translator()


target_language = 'es'


def recognize_speech():
    with sr.Microphone() as source:
        print("Say something...")
        recognizer.adjust_for_ambient_noise(source)
        audio = recognizer.listen(source)
        
        try:
            
            text = recognizer.recognize_google(audio)
            print(f"Original Speech: {text}")
            return text
        except sr.UnknownValueError:
            print("Could not understand audio")
        except sr.RequestError:
            print("Error with the Google Speech Recognition service")
        return None

def text_to_speech_auto(text):
    language = detect(text)
    tts = gTTS(text=text, lang=language)
    
    with io.BytesIO() as audio_buffer:
        tts.write_to_fp(audio_buffer)
        audio_buffer.seek(0)
        
        audio_segment = AudioSegment.from_file(audio_buffer, format="mp3")
        
        print(f"Detected Language: {language}")
        
        play(audio_segment)


# Function to translate recognized speech
def translate_speech(text, target_language):
    try:
        translation = translator.translate(text, dest=target_language)
        print(f"Translated Speech: {translation.text}")
        text_to_speech_auto(translation.text)
    except Exception as e:
        print(f"Error in translation: {e}")

while True:
    speech_text = recognize_speech()
    if speech_text:
        translate_speech(speech_text, target_language)
```
